## Data Preparation 

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(subset="Address", keep = 'first', inplace = True)
df.head(6)
```

### Mengganti Missing Value dengan nilai rata - rata

```python
df.isna().sum()
```

```python
rata_car = df["Car"].mean()
rata_car
```

```python
df["Car"] = df['Car'].fillna(rata_car)
df.head(6)
```

```python
rata_buildingarea = df["BuildingArea"].mean()
rata_buildingarea
```

```python
df["BuildingArea"] = df['BuildingArea'].fillna(rata_buildingarea)
df.head(6)
```

```python
rata_yearbuilt = df["YearBuilt"].mean()
rata_yearbuilt
```

```python
df["YearBuilt"] = df['YearBuilt'].fillna(rata_yearbuilt)
df.head(6)
```

## Menghapus baris dengan yang mengandung missing value

```python
# Menghapus baris dengan nilai yang hilang
df.dropna(subset=['CouncilArea'], inplace=True)

# Menghapus kolom dengan nilai yang hilang
df.dropna(axis=1, inplace=True)
```
