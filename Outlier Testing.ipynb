## Deteksi outlier

#### Data outlier disebabkan karena adanya Variabillitas pada datanya dan terdapat kesalahan pengukuran.
Dampak data outlier :
1. menyebabkan masalah selama melakukan analisis statistik
2. Memberikan dampak yang signifikan terhadap rata rata dan standart deviasinya


#### Untuk mendeteksi data oulier kita dapat melakukan beberapa teknik antara lain :
1. Menggunakan Scatterplot
2. Box Plot
3. Menggunakan Z score
4. Menggunakan nilai IQR 

#### syarat : tidak semua kasus, kita harus merevome data outlier karena kita perlu
1. mempertimbangkan apakah outlier merupakan data eror -> kita dapat meremove
2. atau sebaran data tersebut memang begitu. -> jangan remove data tersebut.
3. terdapat data yang sensitif terhadap outlier, maka model tersebut dapat berkurang performanya

Jika tidak ingin meremove kita bisa melakukan centering dan scaling pada datanya agar meminimalisir dampak oulier tersebut terhadap model.

```python
import matplotlib.pyplot as plt

# Loop untuk membuat histogram untuk setiap variabel
for column in dfnorm.columns:
    # Code to create histogram
    fig, ax = plt.subplots(figsize=(4, 3))  # Menggunakan plt.subplots() dan memperbaiki figsize
    ax.hist(dfnorm[column], bins=100)
    plt.title(f'Histogram of {column}')  # Menambahkan judul histogram
    plt.xlabel(column)  # Label sumbu x sesuai dengan nama variabel
    plt.ylabel('Frequency')  # Label sumbu y

    # Misalkan Anda ingin menambahkan garis vertikal pada nilai tertentu
    limit_atas = dfnorm[column].mean() + 3 * dfnorm[column].std()
    limit_bawah = dfnorm[column].mean() - 3 * dfnorm[column].std()
    ax.axvline(limit_atas, color='r', linestyle='--', label='Upper Limit')
    ax.axvline(limit_bawah, color='r', linestyle='--', label='Lower Limit')
    plt.legend()  # Menambahkan legenda
    plt.show()
```

Terlihat pada visualisasinya terdapat data outlier pada variabel 'Var 1', 'Var 2', 'Var 8', 'Var 9'. Karena terlihat jika terdapat data yang lebih dari limit atas dengan menunjukkan terdapat data yang lebih dari garis putus - putus berwarna merah, sehingga data tersebut dipastikan merupakan data outlier. Dan data yang ada di dalam garis putus 0 putus merah merupakan data normal.

#### Untuk mengetahui terdapat data apa saja yang terdapat outlier dan nilainya kita melakukan deteksi outlier seperti dibawah ini dengan menggunakan Z_Score

```python
import pandas as pd
import numpy as np

def detect_outliers(series, threshold=3):
    outliers = []
    mean = np.mean(series)
    std = np.std(series)
    for value in series:
        z_score = (value - mean) / std 
        if np.abs(z_score) > threshold:
            outliers.append(value)
    return outliers

# Mendeteksi outlier untuk setiap variabel
for column in dfnorm.columns:
    outliers = detect_outliers(dfnorm[column])
    print("Outliers in", column, ":", outliers)
```

terlihat jika terdapat 4 variabel terdapat outlier dengan total outlier sebanyak 8 buah.

Untuk mengatasi adanya outlier ada 3 hal yang dapat dilakukan :
1. menghapus atau membuang data outlier tersebut
2. Melakukan pembatasan pada data ekstrem
3. Melakukan transformasi pada variabelnya

#### Melakukan penanganan Menggunakan Z Score Metode, Karena metodenya sangat cocok digunakan untuk menghapus outlier dengan tipe data berdistribusi normal sehingga dapat mengeluarkan atau meremove data yang tidak sesuai dengan pola lainnya.

```python
limit_atas = dfnorm['Var 9'].mean() + 3*dfnorm['Var 9'].std()
limit_bawah = dfnorm['Var 9'].mean() - 3*dfnorm['Var 9'].std()
print('limit atas: ', limit_atas)
print('limit bawah: ', limit_bawah)
```

```python
dfnorm.loc[(dfnorm['Var 9'] > limit_atas) & (dfnorm['Var 9'] > limit_bawah)]
```

```python
# Deteksi outlier
outliers = dfnorm.loc[(dfnorm['Var 9'] > limit_atas) | (dfnorm['Var 9'] < limit_bawah)].index
# Hapus outlier dari dataframe
dfnorm_cleaned = dfnorm.drop(outliers)
dfnorm_cleaned
```

#### Mendeteksi outlier dengan nilai IQR dengan menunjukkan totalmasing - masing variabel yang memiliki outlier

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membuat function untuk mendeteksi outliner menggunanakan IQR
def detect_outliers_iqr(data, var):
    q1 = data[var].quantile(0.25)
    q3 = data[var].quantile(0.75)
    iqr = q3 - q1 
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = data[(dfnorm_cleaned[var] < lower_bound) | (dfnorm_cleaned [var] > upper_bound)]
    return outliers

# Mendeteksi outlier untuk setiap variabel kontinu
outliers_var1 = detect_outliers_iqr(dfnorm_cleaned, 'Var 1')
outliers_var2 = detect_outliers_iqr(dfnorm_cleaned, 'Var 2')
outliers_var3 = detect_outliers_iqr(dfnorm_cleaned, 'Var 3')
outliers_var4 = detect_outliers_iqr(dfnorm_cleaned, 'Var 4')
outliers_var5 = detect_outliers_iqr(dfnorm_cleaned, 'Var 5')
outliers_var6 = detect_outliers_iqr(dfnorm_cleaned, 'Var 6')
outliers_var7 = detect_outliers_iqr(dfnorm_cleaned, 'Var 7')
outliers_var8 = detect_outliers_iqr(dfnorm_cleaned, 'Var 8')
outliers_var10 = detect_outliers_iqr(dfnorm_cleaned, 'Var 10')

# Printout jumlah outlier yang terdeteksi untuk setiap variabel
print(f"Jumlah outlier yang terdeteksi untuk Var 1: {len(outliers_var1)}")
print(f"Jumlah outlier yang terdeteksi untuk Var 2: {len(outliers_var2)}")
print(f"Jumlah outlier yang terdeteksi untuk Var 3: {len(outliers_var3)}")
print(f"Jumlah outlier yang terdeteksi untuk Var 4: {len(outliers_var4)}")
print(f"Jumlah outlier yang terdeteksi untuk Var 5: {len(outliers_var5)}")
print(f"Jumlah outlier yang terdeteksi untuk Var 6: {len(outliers_var6)}")
print(f"Jumlah outlier yang terdeteksi untuk Var 7: {len(outliers_var7)}")
print(f"Jumlah outlier yang terdeteksi untuk Var 8: {len(outliers_var8)}")
print(f"Jumlah outlier yang terdeteksi untuk Var 10: {len(outliers_var10)}")
```

#### Kemudian melakukan penghapusan pada data outlier dengan metode IQR, karena digunakan untuk data berdistribusi normal dan mampu menghapus data yang jauh dengan rata rata dari data lainnya. 

```python
import numpy as np

# Menghitung kuartil dan IQR untuk Variabel 1, 2, dan 8
Q1 = dfnorm_cleaned[['Var 1', 'Var 2', 'Var 8']].quantile(0.25)
Q3 = dfnorm_cleaned[['Var 1', 'Var 2', 'Var 8']].quantile(0.75)
IQR = Q3 - Q1

# Menghitung batas atas dan bawah
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR

# Menyaring baris-baris di mana variabel 1, 2, dan 8 tidak melebihi batas atas atau bawah
df_out = dfnorm_cleaned[~((dfnorm_cleaned[['Var 1', 'Var 2', 'Var 8']] < lower) | (dfnorm_cleaned[['Var 1', 'Var 2', 'Var 8']] > upper)).any(axis=1)]
df_out
```
